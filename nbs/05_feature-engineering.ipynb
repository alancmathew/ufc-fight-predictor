```python
import numpy as np
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
```

```python
filepath = "../data/clean/completed_fights.pkl"
df = pd.read_pickle(filepath)
```

```python
df.shape
```

```python
df.head()
```

```python
to_drop = [col for col in df.columns if ("fighter" in col) and any([word in col for word in ["overall", "round1", "round2", "round3", "round4", "round5"]])]
```

```python
df = df.drop(columns=to_drop)
```

```python
df
```

```python
fighter1_cols = [col for col in df.columns if "fighter1" in col]
fighter2_cols = [col for col in df.columns if "fighter2" in col]
```

```python
df2 = df.copy(deep=True)
```

```python
df2[fighter1_cols] = df[fighter2_cols]
```

```python
df2[fighter2_cols] = df[fighter1_cols]
```

```python
df = pd.concat([df, df2])
```

```python
df["winner"] = (df["fighter2_status"] == "W").astype("uint8")
```

```python
df
```

```python
tmp = df.select_dtypes("number")
```

```python
tmp
```

```python
from sklearn.metrics import make_scorer, accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
X, y = tmp.drop(columns=["winner"]), tmp["winner"]
```

```python
X = X.fillna(method="backfill")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
accuracy_scorer = make_scorer(accuracy_score)
```

```python
np.mean(cross_val_score(lr, X, y, scoring=accuracy_scorer, cv=5, n_jobs=4, verbose=10))
```

```python
confusion_matrix(y_test, y_pred)
```
